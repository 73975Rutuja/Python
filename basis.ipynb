* BRANCHING (if/ elif/ else):
•	example on it if statements
•	Nested if statements.


```python
# Code 1: Write a code to check if the number is even or odd?

num = int(input("Please enter any number you wanted to check: "))

if num % 2 == 0:
    print(f"Entered number {num} is an Even number")
else:
    print(f"Entered number {num} is an Odd number")
```

```python
# Code 3: Understanding of nested if-else statements:

print(" Simple nested if else statement example")

number1 = int(input("Please enter value for number1 "))

if number1 > 10:
    print("number 1 is greater than 10")
    
    if number1 <= 20:
        print("number 1 is between 10 and 20")
    
    elif number1 <= 50:
        print("number 1 is between 10 and 50")
    
    else:
        print("number 1 is greater than 10, 20, 50")
    
else:
    print("number 1 is less than 10 or equal to 10")
```
